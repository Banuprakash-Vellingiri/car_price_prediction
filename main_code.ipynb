## ___Step 1 : Data Cleaning___

__1. Removing Duplicated Entries :__

```python
df.drop_duplicates(inplace=True)
print("- Duplicated Rows :", df.duplicated().sum())
print("- Shape of the data after removing duplicated values :",df.shape)
```

__2. Removing 'Null' Values :__

```python
df.dropna(inplace=True)
print("Null Values :")
pd.DataFrame(df.isna().sum()).transpose()
```

__3. Cleaning "company" column :__

```python
#Converting the company names into lower case
df["company"]=df["company"].str.lower()
pd.DataFrame(df["company"].value_counts()).transpose()
```

```python
print(df["company"].unique().tolist())
```

```python
#Renaming the company from 'land' to 'land rover'
df["company"]=df["company"].replace({"land":"land rover"})
print("Shape of the data after cleaning 'company' column :",df.shape)
```

__4. Cleaning "year" column :__

```python
#year
pd.DataFrame(df["year"].value_counts()).transpose()
```

```python
print("Data type of 'year' :",df["year"].dtype)
```

```python
#Converting the 'year' column from object data type to numeric(int) data type
df["year"]=pd.to_numeric(df["year"])
#-------------------------------------------------
print("Data type of 'year' :",df["year"].dtype)
```

__5. Cleaning "Price" column :__

```python
print("Data type of 'Price' :",df["Price"].dtype)
```

```python
#Removing commas
df["Price"]=df["Price"].str.replace(",","")
#-------------------------------------------------------------------------------------------------
#Converting the 'Price' column from object data type to numeric data type
df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
#-------------------------------------------------------------------------------------------------
#Since 'NaN' values are replaced in the position where the errors occured in datatype conversion and so the NaN values are removed.
df.dropna(subset=["Price"], inplace=True)
print("Null Values in 'Price' :",df["Price"].isna().sum())
#-------------------------------------------------------------------------------------------------
print("Shape of the data after cleaning 'Price' column :",df.shape)
```

__5. Cleaning "kms_driven" column :__

```python
df.head()
```

```python
#Removing commas
df["kms_driven"]=df["kms_driven"].str.replace(",","")
#Removing "kms"
df["kms_driven"]=df["kms_driven"].str.replace("kms","")
#-------------------------------------------------------------------------------------------------
#Converting the 'Price' column from object data type to numeric data type
df["kms_driven"]=pd.to_numeric(df["kms_driven"],errors="coerce")
#-------------------------------------------------------------------------------------------------
#Since NaN values are replaced in the position where the errors occured in datatype conversion and so the NaN values are removed.
df.dropna(subset=["kms_driven"], inplace=True)
print("Null Values in 'kms_driven' :",df["kms_driven"].isna().sum())
#-------------------------------------------------------------------------------------------------
print("Shape of the data after cleaning 'kms_driven' column :",df.shape)
```

__6. Cleaning "name" column :__

```python
#Extrcating name
def extract_name(x):
    x=x.lower()
    x=x.split()
    x=x[:3]
    return(" ".join(map(str,x)))
df['name']=df['name'].apply(lambda x: extract_name(x) )
#-------------------------------------------------------------------------------------------------
print("Shape of the data after cleaning 'name' column :",df.shape)
```

__6. Cleaning "fuel_type" column :__

```python
#Converting "fuel_type" column categories to lower case for convinience
df["fuel_type"]=df["fuel_type"].str.lower()
```

#### ___Data Preprocessing Completed âœ…___
- `The dataset has been cleaned to a satisfactory standard and is now suitable for training machine learning models.`

## ___Step 3 : Machine Learning___

___Model Selection :___

- _Four different models, including_

     `1. Linear Regression`
     
     `2. Decision Tree Regressor`

    ` 3. Random Forest Regressor`
    
     `4. Support Vector Classifier`

_have been trained and evaluated for their performance._

```python
#Importing Dependencies
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

___Splitting the datasets for training and testing :___

```python
df.reset_index(inplace=True)
#Feature variables(X)
X=df[['name', 'company', 'year','kms_driven', 'fuel_type']]
#---------------------------------------------
#Target Variable(y)
y=df["Price"]
```

```python
X["year"].head(5)
```

___One Hot Encoding :___
- Categorical columns are encoded using OHE.

```python
ohe_encoder = OneHotEncoder()
encoded_data= ohe_encoder.fit_transform(X[['name', 'company', 'fuel_type']]).toarray()
#---------------------------------------------
X_encoded= pd.DataFrame(encoded_data, columns=ohe_encoder.get_feature_names_out(['name', 'company', 'fuel_type']))
#---------------------------------------------
#Adding year column to the new encoded X dataframe
X_encoded['year'] = X['year']
#---------------------------------------------
X_encoded.head(5)
```

```python
X_encoded= pd.get_dummies(df, columns=['name', 'company', 'fuel_type'])
```

___Train Test Split :___

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

___Scaling the data using Standard Scalar :___

```python
sc = StandardScaler()
#---------------------------------------------
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
models = {
            "Linear Regression": LinearRegression(),
            "Decision Tree": DecisionTreeRegressor(),
            "Random Forest": RandomForestRegressor(),
            "SVR": SVR(),
         }
#----------------------------------------------------------------------------------
results = []
#----------------------------------------------------------------------------------
for name, model in models.items():
    #----------------------------------------------------------------------------------
    model.fit(X_train_scaled, y_train)
    #----------------------------------------------------------------------------------
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)
    #----------------------------------------------------------------------------------
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    #----------------------------------------------------------------------------------
    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    #----------------------------------------------------------------------------------
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    #----------------------------------------------------------------------------------
    results.append({'Model': name, 'Train MSE': train_mse, 'Test MSE': test_mse, 
                    'Train MAE': train_mae, 'Test MAE': test_mae, 
                    'Train R2 Score': train_r2, 'Test R2 Score': test_r2})
#----------------------------------------------------------------------------------   
metrics_df = pd.DataFrame(results)
metrics_df
```

```python
#Plotting Graph
plt.figure(figsize=(20, 4))
#----------------------------------------------------------------------------------
# Train MSE
plt.plot(metrics_df['Model'],metrics_df['Train MSE'], marker='o', label='Train MSE')
#----------------------------------------------------------------------------------
# Test MSE
plt.plot(metrics_df['Model'], metrics_df['Test MSE'], marker='o', label='Test MSE')
#----------------------------------------------------------------------------------
plt.title('Train and Test MSE Comparison')
plt.xlabel('Model')
plt.ylabel('MSE')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
___Observations :___

- _The linear regression model shows excellent performance on the training set with a good R2 score, but very lower performance on the test compared to the training set, indicating overfitting._
- _Decision tree model performs very well in training with zero MSE and in case of testing its MSE is too high which explicitly indicates too much overfitted ._
- Random Forest performs good when compared to other models whoich performs good both in testing and training . 
- _SVR performs poorly compared to other models, with significantly higher MSE and MAE values and negative R2 scores, indicating poor fit and possibly inappropriate model choice for the data._


`Out of the 4 models evaluated, the "Random Forest Regressor" displayed the most consistent and effective performance across both training and testing phases.`

### ___Random Forest Model With Hyper Parameter Tuning :___

- _Using "RandomizedSearchCV" for hyper parameter tuning._

```python
#Importing Dependencies 
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
#----------------------------------------------------------------------------------
rf_model = RandomForestRegressor()
random_search= RandomizedSearchCV(estimator = rf_model , param_distributions = param_grid, 
                               n_iter = 100, cv = 3, verbose = 2, random_state = 42, n_jobs = -1)
#----------------------------------------------------------------------------------
random_search.fit(X_train_scaled, y_train)
best_params = random_search.best_params_
best_rf_model = RandomForestRegressor(**best_params)
best_rf_model.fit(X_train_scaled, y_train)
y_pred=best_rf_model.predict(X_test_scaled)
#----------------------------------------------------------------------------------
print("Best parameters:",best_params)
```

___Actual Output vs Predicted Output :___

```python
result_df= pd.DataFrame({'Actual Outcome': y_test, 'Predicted Outcome': y_pred})
result_df['Predicted Outcome'] = result_df['Predicted Outcome'].astype(str).astype(float).round()
result_df.sample(n=7)
```

```python
#Plotting Graph
plt.figure(figsize=(20, 4))
plt.scatter(result_df['Actual Outcome'], result_df['Predicted Outcome'], color='red')
plt.plot(result_df['Actual Outcome'], result_df['Actual Outcome'], color='blue',)
plt.title('Actual Outcome vs Predicted Outcome')
plt.xlabel('Actual Outcome')
plt.ylabel('Predicted Outcome')
plt.grid(True)
plt.show()
```
